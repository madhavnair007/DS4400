```python
import pandas as pd
```

```python
df = pd.read_csv("kc_house_data.csv")
df = df.drop(columns=["id", "date", "zipcode"])
corr = df.corr(numeric_only=True)["price"].drop("price").sort_values(ascending=False)

print(corr)
```

```python
positive_features = corr[corr > 0].index.tolist()
print("\nPositively correlated features:", positive_features)

top_feature = corr.idxmax()
top_value = corr.max()
print("\nHighest positive correlation:", top_feature, top_value)
```
