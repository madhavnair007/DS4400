```python
import numpy as np
```

```python
# np.random.seed(0)  # optional, for reproducibility

A1 = np.random.randint(-10, 11, size=(3, 3))
A2 = np.random.randint(-10, 11, size=(3, 3))
A3 = np.random.randint(-10, 11, size=(3, 3))

A1, A2, A3
```

```python
np.linalg.det(A1)
```

```python
A1_inv = np.linalg.inv(A1)
A1_inv
```

```python
A2_inv = np.linalg.inv(A2)
A2_inv
```

```python
A3_inv = np.linalg.inv(A3)
A3_inv
```

```python
A1 @ A1_inv
```

```python
A2 @ A2_inv
```

```python
A3 @ A3_inv
```
