```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("kc_house_data.csv")

# Drop columns to ignore
df = df.drop(columns=["id", "date", "zipcode"])

# Compute statistics for each feature
stats = df.agg(["mean", "min", "max", "var"]).T

# Display full statistics table
stats
```

```python
# Lowest and highest mean
lowest_mean_feature = stats["mean"].idxmin()
lowest_mean_value = stats.loc[lowest_mean_feature, "mean"]

highest_mean_feature = stats["mean"].idxmax()
highest_mean_value = stats.loc[highest_mean_feature, "mean"]

# Lowest and highest variance
lowest_var_feature = stats["var"].idxmin()
lowest_var_value = stats.loc[lowest_var_feature, "var"]

highest_var_feature = stats["var"].idxmax()
highest_var_value = stats.loc[highest_var_feature, "var"]

print("Lowest mean:", lowest_mean_feature, lowest_mean_value)
print("Highest mean:", highest_mean_feature, highest_mean_value)
print("Lowest variance:", lowest_var_feature, lowest_var_value)
print("Highest variance:", highest_var_feature, highest_var_value)
```
