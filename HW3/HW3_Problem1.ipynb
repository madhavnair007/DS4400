# Part 1

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
df = pd.read_csv("spambase.data", header=None)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
model = LogisticRegression(max_iter=3000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
accuracy = accuracy_score(y_test, y_pred)
error = 1 - accuracy

print("Accuracy:", accuracy)
print("Error:", error)
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

# Part 2

```python
coefficients = model.coef_[0]

for i, coef in enumerate(coefficients):
    print(f"Feature {i}: {coef}")
```

```python
feature_importance = pd.DataFrame({
    "Feature": range(len(coefficients)),
    "Coefficient": coefficients,
    "Abs": np.abs(coefficients)
})

feature_importance = feature_importance.sort_values(by="Abs", ascending=False)

print(feature_importance.head(10))
```

```python
positive = feature_importance[feature_importance["Coefficient"] > 0]
negative = feature_importance[feature_importance["Coefficient"] < 0]

print("Top Positive Features:")
print(positive.head(5))

print("\nTop Negative Features:")
print(negative.head(5))
```

# Part 3

```python
y_probs = model.predict_proba(X_test)[:, 1]

thresholds = [0.25, 0.5, 0.75, 0.9]

results = []

for T in thresholds:
    y_pred_T = (y_probs >= T).astype(int)
    
    acc = accuracy_score(y_test, y_pred_T)
    prec = precision_score(y_test, y_pred_T)
    rec = recall_score(y_test, y_pred_T)
    
    results.append((T, acc, prec, rec))

for r in results:
    print(f"T={r[0]} | Accuracy={r[1]} | Precision={r[2]} | Recall={r[3]}")
```
