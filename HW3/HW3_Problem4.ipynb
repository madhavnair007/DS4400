# Part 1

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle

df = pd.read_csv("spambase.data", header=None)

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

X, y = shuffle(X, y, random_state=42)

k = 5

def manual_k_fold_cv(model_class, X, y, k):
    n = len(X)
    fold_size = n // k
    errors = []

    for i in range(k):
        start = i * fold_size
        end = (i + 1) * fold_size if i != k - 1 else n

        X_val = X[start:end]
        y_val = y[start:end]

        X_train = np.concatenate((X[:start], X[end:]), axis=0)
        y_train = np.concatenate((y[:start], y[end:]), axis=0)

        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_val_scaled = scaler.transform(X_val)

        model = model_class()
        if model_class == LogisticRegression:
            model = model_class(max_iter=1000)

        model.fit(X_train_scaled, y_train)

        y_val_pred = model.predict(X_val_scaled)

        acc = accuracy_score(y_val, y_val_pred)
        error = 1 - acc
        errors.append(error)

        print(f"Fold {i+1}: Validation Error = {error:.4f}")

    avg_error = np.mean(errors)
    print(f"Average Validation Error = {avg_error:.4f}")
    return errors, avg_error

print("Logistic Regression CV Results")
log_errors, log_avg_error = manual_k_fold_cv(LogisticRegression, X, y, k)

print("\nLDA CV Results")
lda_errors, lda_avg_error = manual_k_fold_cv(LinearDiscriminantAnalysis, X, y, k)
```

```python
def manual_k_fold_cv(model_class, X, y, k):
    n = len(X)
    fold_size = n // k
    errors = []

    for i in range(k):
        start = i * fold_size
        end = (i + 1) * fold_size if i != k - 1 else n

        X_val = X[start:end]
        y_val = y[start:end]

        X_train = np.concatenate((X[:start], X[end:]), axis=0)
        y_train = np.concatenate((y[:start], y[end:]), axis=0)

        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_val_scaled = scaler.transform(X_val)

        if model_class == LogisticRegression:
            model = model_class(max_iter=1000)
        else:
            model = model_class()

        model.fit(X_train_scaled, y_train)
        y_val_pred = model.predict(X_val_scaled)

        acc = accuracy_score(y_val, y_val_pred)
        error = 1 - acc
        errors.append(error)

    avg_error = np.mean(errors)
    return avg_error

k_values = [5, 10]

print("Logistic Regression")
for k in k_values:
    avg_error = manual_k_fold_cv(LogisticRegression, X, y, k)
    print(f"k={k} | Average Validation Error = {avg_error:.4f}")

print("\nLDA")
for k in k_values:
    avg_error = manual_k_fold_cv(LinearDiscriminantAnalysis, X, y, k)
    print(f"k={k} | Average Validation Error = {avg_error:.4f}")
```
